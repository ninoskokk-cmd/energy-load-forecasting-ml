```python
df_weather_pivot=df_weather.pivot(columns='city_name',index='dt_iso',values=['temp', 'pressure', 'humidity', 'wind_speed', 'wind_deg', 'rain_1h', 'snow_3h', 'clouds_all'])
```

```python
df_weather_pivot.head()
```

```python
new_columns=[]
for col in df_weather_pivot.columns:
    combined_name=col[0]+ "_" + col[1]
    new_columns.append(combined_name)

df_weather_pivot.columns=new_columns
    
```

```python
df_weather_pivot.info()
```

```python
df_final=pd.merge(df_energy,df_weather_pivot,left_on='time',right_on='dt_iso')
```

```python
df_final.info()
```

```python
df_final['year']=df_final['time'].apply(lambda time:time.year)
df_final['month']=df_final['time'].apply(lambda time:time.month)
df_final['day']=df_final['time'].apply(lambda time:time.day)
df_final['hour']=df_final['time'].apply(lambda time:time.hour)
```

```python
df_final.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df_final.drop(['time','total load actual'],axis=1)
y=df_final['total load actual']
train_data=df_final[df_final['year']<2018]
test_data=df_final[df_final['year']==2018]
X_train=train_data.drop(['time', 'total load actual'], axis=1)
y_train=train_data['total load actual']
X_test=test_data.drop(['time', 'total load actual'], axis=1)
y_test=test_data['total load actual']
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor()
```

```python
rfr.fit(X_train,y_train)
```

```python
predictions=rfr.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test,predictions))

print('MSE:', metrics.mean_squared_error(y_test,predictions))

print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

```python
plt.figure(figsize=(15,6))
plt.plot(y_test.values[:336], label='Real Demand', color='royalblue', linewidth=2) 
plt.plot(predictions[:336], label='RandomForest prediction', color='tomato', linestyle='--', linewidth=2)
plt.title('Comparison of Forecast and Actual Demand (January 2018)', fontsize=14)
plt.xlabel('Hours', fontsize=12)
plt.ylabel('MW', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
